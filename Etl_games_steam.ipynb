```python
#Cargamos las librerías que  vayamos a utilizar
import pandas as pd
import json
import ast
import pyarrow as pa
```

```python

#ruta del archivo steam games
ruta = 'C://Users//CRISTIAN//Desktop//Proyecto final #  1//Proyecto-Final-individual-1-Henry//datasets//output_steam_games.json'

#aqui se almacenan los diccionarios generados a partir de cada linea del archivo
lista = []

#se  abre el archivo y se lee línea por línea, como un archivo Json
with open(ruta, 'r') as archivo:
        for linea in archivo:
            try:
                diccionario_datos = json.loads(linea)
                lista.append(diccionario_datos)
            except json.JSONDecodeError as i:
                print(f"Error de decodificación JSON en línea: {i}")
#Se crea un Dataframe apartir de la lista de un diccionario
games_steam_df = pd.DataFrame(lista)
#Mostramos el dataframe
games_steam_df
```

Transformaciones en la columna 'price'

```python
#Visualizamos los valores únicos
games_steam_df['price'].unique()
```

```python
#Reemplazamos los valores de tipo nulo por '--'
games_steam_df['price'] = games_steam_df['price'].fillna('--')
```

```python
#Vemos los valores que no son numericos y los guardamos en 'lista'
patron = r'^\D*$' #Patron para extraer los valores que no sean numericos

lista = games_steam_df[games_steam_df['price'].str.contains(patron, case=False, regex=True)]['price'].unique() #Extraemos los valores que no sean numericos
```

```python
lista_gratis = []

for i in lista:
    if i != '--':
        lista_gratis.append(i)
else:
    pass
```

```python
#Reemplazamos esos valores, por el valor '0.0'
games_steam_df['price'] = games_steam_df['price'].replace(lista, '0.0')
```

```python
games_steam_df['price'].unique()
```

```python
#Hay valores con texto, pero que tambien tienen un precio, asique a esos, solo le dejaremos el precio
games_steam_df['price'] = games_steam_df['price'].replace('Starting at $499.00', '499.00')
```

```python
#Cambiamos el tipo de dato de la columna 'price' a float
games_steam_df['price'] = games_steam_df['price'].astype(float)
```

```python
#verificamos
games_steam_df['price'].dtype
```

Transformaciones en la columna 'genres'

```python
#Descomponemos la columna 'genres', ya que tiene valores agrupados en listas
games_steam_df['genres'] = games_steam_df['genres'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])

games_steam_df = games_steam_df.drop('genres', axis=1).join(games_steam_df['genres'].apply(pd.Series).stack().reset_index(level=1, drop=True).rename('genres'))
```

```python
#Verificamos que los valores no esten agrupados en listas
games_steam_df['genres'].unique()
```

```python
#Hay un error al identificar '&', asique lo corregimos
games_steam_df['genres'] = games_steam_df['genres'].str.replace('&amp;', 'and')
```
