```python
#Cargamos las librerías que  vayamos a utilizar
import pandas as pd
import json
import ast
import pyarrow as pa
import numpy as np
```

```python
#ruta del archivo steam games
ruta = './/datasets//australian_user_reviews.json'
#Se extraen los caracteres que nos son apropiados de un JSON y se reemplazan por los apropiados
data = []
patrones = [ 
["\"","TEMP@TEMP"],
["{'","{\""],
["'}","\"}"],
["': '","\": \""],
["', '","\", \""],
["': \"","\": \""],
["': ","\": "],
["True, '","true, \""],
["False, '","false, \""],
["\\'","'"],
["\\","\\\\"],
["\": TEMP@TEMP","\": \""],   
["TEMP@TEMP}","\"}"],   
["TEMP@TEMP","'"],   
["e\", \"T","e','T"]
]
#convertimos el json a un json valido, extraemos los datos de la columna reviews y los listamos en un dataset aparte apuntando al 'user_id' especifico de cada registro principal
#de tal forma que en este punto contamos con dos listados de datos para posteriormente mezclarlos y formar un solo dataset en toda la data
reviewsJsonAux: any
j=0
with open(ruta, 'r', encoding='utf8') as archivo:
    for linea in archivo:
        try:
            for i in range(0,np.shape(patrones)[0]):
                linea = linea.replace(patrones[i][0],patrones[i][1])            
            diccionario_datos = json.loads(linea)
            data.append(diccionario_datos)
            strReviews = json.dumps(diccionario_datos['reviews']).replace("\"funny\":","\"user_id\":\""+diccionario_datos['user_id']+"\",\"funny\":")
            reviewsJson = json.loads(strReviews)
            if j == 0 :
                j+=1
                reviewsJsonAux = reviewsJson
            else:
                reviewsJsonAux  += reviewsJson
        except json.JSONDecodeError as i:
            a=1

#Se crea un Dataframe apartir de la lista de un diccionario
user_reviews_df = pd.DataFrame(data)
user_reviews_dfa = pd.DataFrame(reviewsJsonAux)
user_reviews_dfb = user_reviews_df.merge(user_reviews_dfa, left_on='user_id', right_on='user_id')
#Mostramos el dataframe
user_reviews_dfb

```
